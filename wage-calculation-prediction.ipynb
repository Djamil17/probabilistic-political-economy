```python
import numpy as np
import pandas as pd

f= '../data/US-Wages.csv'

with open(f, 'r'):
    
    data=pd.read_csv(f)

data=data.dropna(how='all')
data.replace('[$#%^@%*]','0',regex=True,inplace=True)
data.replace(',','0',regex=True,inplace=True)

nums=[ 'tot_emp', 'emp_prse',
       'jobs_1000', 'loc_quotient', 'pct_total', 'h_mean', 'a_mean',
       'mean_prse', 'h_pct10', 'h_pct25', 'h_median', 'h_pct75', 'h_pct90',
       'a_pct10', 'a_pct25', 'a_median', 'a_pct75', 'a_pct90', 'annual',
       'hourly']

data[nums]=data[nums].astype(np.float64)
subset=data[data.o_group=='detailed']
subset
```

```python
import plotly.express as px

shape = subset['h_median'].shape[0]
bins = int(np.round(np.sqrt(shape)))

px.histogram(subset,
             'h_median',
             'tot_emp',
             histnorm='probability',
             nbins=bins,
             title='Hourly Median Wage',
             color_discrete_sequence=["magenta"],
                         opacity=.7,
             labels={
                 "h_median": "Hourly Median $",
                 "tot_emp": "Probability"
             })

```

```python
shape = subset['h_median'].shape[0]
bins = int(np.round(np.sqrt(shape)))

px.histogram(subset,
             'h_mean',
             'tot_emp',
             histnorm='probability',
             nbins=bins,
             title='Hourly Mean Wage',
             color_discrete_sequence=["yellow"],
                         opacity=.7,
             labels={
                 "h_median": "Hourly Median $",
                 "tot_emp": "Probability"
             })
```
